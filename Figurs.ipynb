```python
import pandas as pd
from plotly import express as px
```

```python
path = r'C:\Users\Владислав\Мой диск\Python\Python\HH.ru\combined_data.xlsx'
df = pd.read_excel(path)
df.head()
```

```python
df = df[(df['(salary)_from_in_RUR'] >= 0) & (df['(salary)_from_in_RUR'] <= 50000)]

max_salary_from_in_RUR = max(df['(salary)_from_in_RUR'])
min_salary_from_in_RUR = min(df['(salary)_from_in_RUR'])
print(max_salary_from_in_RUR, min_salary_from_in_RUR)
df.columns
```

```python
agg_func_count = {'(salary)_from_in_RUR': ['mean']}
```

```python
grouped_data = df.groupby('(employer)_name', as_index=False).agg(agg_func_count).round(2)
grouped_data.head()
grouped_data.columns = ['_'.join(col).rstrip('_') for col in grouped_data.columns.values]
grouped_data = grouped_data.sort_values(by='(salary)_from_in_RUR_mean', ascending=False)
grouped_data = grouped_data.head(30)
grouped_data = grouped_data.sort_values(by='(salary)_from_in_RUR_mean', ascending=True)
```

```python
fig = px.bar(grouped_data, y='(employer)_name', x='(salary)_from_in_RUR_mean', title='Количество значений по категориям', orientation = 'h')

# Отображение диаграммы
fig.show()
```

```python
df_filtered = df.dropna(subset=['(address)_lat', '(address)_lng'])
fig = px.scatter_mapbox(df_filtered, lat = '(address)_lat', lon = '(address)_lng', color='(salary)_from_in_RUR', size='(salary)_from_in_RUR', size_max = 10, zoom=3, height=300)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.update_layout(mapbox_style="open-street-map", margin={"r": 0, "t": 0, "l": 0, "b": 0})
fig.show()


#fig = px.scatter_mapbox(us_cities, lat="lat", lon="lon", hover_name="City", hover_data=["State", "Population"],
#                        color_discrete_sequence=["fuchsia"], zoom=3, height=300)
```
